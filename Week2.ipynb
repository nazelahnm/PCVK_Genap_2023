```python
black_img2 = np.zeros(shape=(512,512,3), dtype=np.int32)
plt.imshow(black_img2)
```

```python
vertices = np.array([[100,300], [200,200], [400,300], [200,400]], dtype=np.int32)
vertices
```

```python
pts = vertices.reshape((-1,1,2)) #Nilai 2 untuk menunjukkan bahwa tiap titik dibuat 3 channel yg mewakili R,G,B
pts
```

```python
#Menambahkan Polyline
cv.polylines(black_img2,[pts],isClosed=True,color=(255,0,0), thickness = 5)
plt.imshow(black_img2)
```

```python
#2. Red-Blue

img_rb = img.copy()

# Split channels
b, g, r = cv.split(img)

#gabung chanel red dan blue
img_rb = cv.merge([r, b, b])

# menampilkan gambar
plt.imshow(img_rb)
plt.show()
```

```python
#3. Green-Blue

img_rb = img.copy()

# Split channels
b, g, r = cv.split(img)

# gabung chanel green dan blue
img_rb = cv.merge([g, b, b])

# menampilkan gambar
plt.imshow(img_rb)
plt.show()
```

```python
#3. Menampilkan image baris ke 20-115, kolom 25-120

imgt = img[20:115,25:120]
plt.imshow(imgt)
```

```python
#4. Menampilkan image baris ke 5-30, semua kolom, channel Red saja

img_rb = img.copy()

# memisahkan saluran warna merah
b, g, r = cv.split(img)
red_channel = r

# membuat citra hanya dengan saluran warna merah
img_red = cv.merge([red_channel, red_channel, red_channel])

# menampilkan citra hanya dengan saluran warna merah
plt.imshow(img_red[5:30], cmap='gray')
plt.show()
```
